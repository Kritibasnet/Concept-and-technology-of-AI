```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the South Asian dataset
south_asia_data = pd.read_csv('south_asian_countries.csv')

# Step 1: Create the Composite Score column
south_asia_data['Composite Score'] = (0.40 * south_asia_data['Log GDP per capita'] +
                                      0.30 * south_asia_data['Social support'] +
                                      0.30 * south_asia_data['Healthy life expectancy'])

# Step 2: Rank the countries based on Composite Score in descending order
south_asia_data = south_asia_data.sort_values(by='Composite Score', ascending=False)

# Step 3: Visualize the top 5 countries using a horizontal bar chart
top_5_composite = south_asia_data.head(5)

plt.figure(figsize=(10, 6))
bars = plt.barh(top_5_composite['Country name'], top_5_composite['Composite Score'], color='forestgreen')
plt.title('Top 5 South Asian Countries by Composite Score', fontsize=16)
plt.xlabel('Composite Score', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Annotate the bar chart with the score values
for bar in bars:
    plt.text(bar.get_width() + 0.02, bar.get_y() + bar.get_height() / 2,
             f'{bar.get_width():.2f}', va='center', ha='left', fontsize=10)

plt.tight_layout()
plt.show()

# Step 4: Compare the rankings with the original Score column
# Sorting the countries by the original Score column
south_asia_data_sorted_by_score = south_asia_data.sort_values(by='score', ascending=False)

# Plotting a comparison chart with a clearer distinction between the two scores
plt.figure(figsize=(10, 6))

# Plot the Composite Score
plt.plot(south_asia_data_sorted_by_score['Country name'],
         south_asia_data_sorted_by_score['Composite Score'],
         label='Composite Score', marker='o', color='darkblue', linestyle='-', linewidth=2)

# Plot the Original Score for comparison
plt.plot(south_asia_data_sorted_by_score['Country name'],
         south_asia_data_sorted_by_score['score'],
         label='Original Score', marker='x', color='darkred', linestyle='--', linewidth=2)

# Add labels and title
plt.title('Comparison of Composite Score vs Original Score', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

Task - 3 - Outlier Detection:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the South Asian dataset
south_asia_data = pd.read_csv('south_asian_countries.csv')

# Step 1: Identify outliers using the 1.5 × IQR rule for 'Score' and 'GDP per Capita'

# Function to calculate IQR and identify outliers
def identify_outliers(df, column):
    Q1, Q3 = df[column].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    return df[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]

# Identify outliers for 'Score' and 'GDP per Capita'
outliers_score = identify_outliers(south_asia_data, 'score')
outliers_gdp = identify_outliers(south_asia_data, 'Log GDP per capita')

# Step 2: Create a scatter plot and highlight outliers
plt.figure(figsize=(10, 6))

# Scatter plot for GDP per Capita vs Score
plt.scatter(south_asia_data['Log GDP per capita'], south_asia_data['score'], color='blue', label='Normal Data', alpha=0.7)

# Highlight outliers in a different color (red) for Score
plt.scatter(outliers_score['Log GDP per capita'], outliers_score['score'], color='red', label='Outliers (Score)', s=100, edgecolor='black')

# Highlight outliers in GDP per Capita (green)
plt.scatter(outliers_gdp['Log GDP per capita'], outliers_gdp['score'], color='green', label='Outliers (GDP)', s=100, edgecolor='black')

# Add labels and title
plt.title('Scatter Plot: GDP per Capita vs Happiness Score with Outliers', fontsize=16)
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 3: Display outliers based on Score and GDP per Capita
print("Outliers based on Score:")
print(outliers_score[['Country name', 'score', 'Log GDP per capita']])

print("\nOutliers based on GDP per Capita:")
print(outliers_gdp[['Country name', 'score', 'Log GDP per capita']])


```

Task - 4 - Exploring Trends Across Metrics:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the South Asian dataset
south_asia_data = pd.read_csv('south_asian_countries.csv')

# Step 1: Calculate the Pearson correlation between selected metrics and the Score
correlation_freedom = south_asia_data['Freedom to make life choices'].corr(south_asia_data['score'])
correlation_generosity = south_asia_data['Generosity'].corr(south_asia_data['score'])

# Print the correlations
print(f"Correlation between Freedom to Make Life Choices and Score: {correlation_freedom:.2f}")
print(f"Correlation between Generosity and Score: {correlation_generosity:.2f}")

# Step 2: Create scatter plots with trendlines

# Scatter plot for Freedom to Make Life Choices vs Score with trendline
plt.figure(figsize=(10, 6))
sns.regplot(x='Freedom to make life choices', y='score', data=south_asia_data, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Freedom to Make Life Choices vs Happiness Score', fontsize=16)
plt.xlabel('Freedom to Make Life Choices', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Scatter plot for Generosity vs Score with trendline
plt.figure(figsize=(10, 6))
sns.regplot(x='Generosity', y='score', data=south_asia_data, scatter_kws={'color':'green'}, line_kws={'color':'red'})
plt.title('Generosity vs Happiness Score', fontsize=16)
plt.xlabel('Generosity', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 3: Identify and discuss the strongest and weakest relationships

# Based on the correlation values printed above:
if abs(correlation_freedom) > abs(correlation_generosity):
    print("The strongest relationship is between Freedom to Make Life Choices and Happiness Score.")
else:
    print("The strongest relationship is between Generosity and Happiness Score.")

if abs(correlation_freedom) < abs(correlation_generosity):
    print("The weakest relationship is between Freedom to Make Life Choices and Happiness Score.")
else:
    print("The weakest relationship is between Generosity and Happiness Score.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the South Asian and Middle Eastern data
south_asia_df = pd.read_csv('south_asian_countries.csv')
middle_east_df = pd.read_csv('middle_eastern_countries.csv')

# Function to identify outliers using the 1.5 × IQR rule
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# 1. Identify outliers for both Score and GDP per Capita in South Asia
south_asia_score_outliers = identify_outliers(south_asia_df, 'score')
south_asia_gdp_outliers = identify_outliers(south_asia_df, 'Log GDP per capita')

# 2. Identify outliers for both Score and GDP per Capita in Middle East
middle_east_score_outliers = identify_outliers(middle_east_df, 'score')
middle_east_gdp_outliers = identify_outliers(middle_east_df, 'Log GDP per capita')

# 3. Plot outliers for South Asia - Score vs GDP per Capita
plt.figure(figsize=(14, 6))

# South Asia - Score vs GDP per Capita
plt.subplot(1, 2, 1)
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], color='blue', label='Normal Data')
plt.scatter(south_asia_score_outliers['Log GDP per capita'], south_asia_score_outliers['score'], color='red', label='Score Outliers', s=100, edgecolor='black')
plt.scatter(south_asia_gdp_outliers['Log GDP per capita'], south_asia_gdp_outliers['score'], color='green', label='GDP Outliers', s=100, edgecolor='black')
plt.title('South Asia: Score vs GDP per Capita with Outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()

# 4. Plot outliers for Middle East - Score vs GDP per Capita
plt.subplot(1, 2, 2)
plt.scatter(middle_east_df['Log GDP per capita'], middle_east_df['score'], color='blue', label='Normal Data')
plt.scatter(middle_east_score_outliers['Log GDP per capita'], middle_east_score_outliers['score'], color='red', label='Score Outliers', s=100, edgecolor='black')
plt.scatter(middle_east_gdp_outliers['Log GDP per capita'], middle_east_gdp_outliers['score'], color='green', label='GDP Outliers', s=100, edgecolor='black')
plt.title('Middle East: Score vs GDP per Capita with Outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()

plt.tight_layout()
plt.show()

# Display outliers for South Asia and Middle East
print("South Asia - Score Outliers:")
print(south_asia_score_outliers[['Country name', 'score', 'Log GDP per capita']])

print("\nSouth Asia - GDP per Capita Outliers:")
print(south_asia_gdp_outliers[['Country name', 'score', 'Log GDP per capita']])

print("\nMiddle East - Score Outliers:")
print(middle_east_score_outliers[['Country name', 'score', 'Log GDP per capita']])

print("\nMiddle East - GDP per Capita Outliers:")
print(middle_east_gdp_outliers[['Country name', 'score', 'Log GDP per capita']])
```
